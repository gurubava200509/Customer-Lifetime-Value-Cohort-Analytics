```python
df_orders_revenue_final["order_purchase_timestamp"].max()
```

```python
df_orders_revenue_final["order_purchase_timestamp"] = pd.to_datetime(
    df_orders_revenue_final["order_purchase_timestamp"]
)
```

```python
snapshot_date = (
    df_orders_revenue_final["order_purchase_timestamp"].max()
    + pd.Timedelta(days=1)
)

snapshot_date
```

```python
rfm = df_orders_revenue_final.groupby("customer_unique_id").agg(
    Recency=("order_purchase_timestamp",
             lambda x: (snapshot_date - x.max()).days),

    Frequency=("order_id", "nunique"),

    Monetary=("payment_value", "sum")
).reset_index()

rfm.head()
```

```python
rfm.describe()
```

```python
rfm["Frequency"].value_counts().sort_index()
```

```python
rfm["R_Score"] = pd.qcut(
    rfm["Recency"],
    5,
    labels=[5, 4, 3, 2, 1]
).astype(int)

rfm["R_Score"].value_counts().sort_index()
```

```python
rfm["M_Score"] = pd.qcut(
    rfm["Monetary"],
    5,
    labels=[1, 2, 3, 4, 5]
).astype(int)

rfm["M_Score"].value_counts().sort_index()
```

```python
rfm["F_Score"] = rfm["Frequency"].apply(
    lambda x: 1 if x == 1
    else 2 if x == 2
    else 3
)

rfm["F_Score"].value_counts().sort_index()
```

```python
rfm["RFM_Score"] = (
    rfm["R_Score"].astype(str) +
    rfm["F_Score"].astype(str) +
    rfm["M_Score"].astype(str)
)

rfm[
    ["customer_unique_id",
     "R_Score",
     "F_Score",
     "M_Score",
     "RFM_Score"]
].head()
```

```python
rfm["RFM_Score"].value_counts().head(20)
```

```python
def segment_customer(row):

    if row["F_Score"] >= 2 and row["R_Score"] >= 4:
        return "Champions"

    elif row["F_Score"] >= 2:
        return "Loyal Customers"

    elif row["R_Score"] >= 4 and row["M_Score"] >= 4:
        return "High Value New"

    elif row["R_Score"] <= 2:
        return "At Risk"

    else:
        return "Regular Customers"


rfm["Segment"] = rfm.apply(segment_customer, axis=1)

rfm["Segment"].value_counts()
```

```python
rfm.groupby("Segment").agg({
    "Recency": "mean",
    "Frequency": "mean",
    "Monetary": "mean"
}).round(2)
```

```python
segment_summary = rfm.groupby("Segment").agg(
    Customers=("customer_unique_id", "count"),
    Avg_Recency=("Recency", "mean"),
    Avg_Frequency=("Frequency", "mean"),
    Avg_Monetary=("Monetary", "mean"),
    Total_Revenue=("Monetary", "sum")
).round(2)

segment_summary
```

```python
segment_summary["Revenue_%"] = (
    segment_summary["Total_Revenue"]
    / segment_summary["Total_Revenue"].sum()
    * 100
).round(2)

segment_summary.sort_values(
    "Revenue_%",
    ascending=False
)
```

```python
segment_summary["Customer_%"] = (
    segment_summary["Customers"]
    / segment_summary["Customers"].sum()
    * 100
).round(2)

segment_summary.sort_values(
    "Customer_%",
    ascending=False
)
```
